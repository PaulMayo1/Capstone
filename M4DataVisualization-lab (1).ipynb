## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt


```

```python
QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

```

```python
df=pd.read_sql_query(QUERY,conn)  
df.set_index('DatabaseDesireNextYear', inplace=True) 

color_list=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

df['count'].plot(kind='pie', figsize=(20,6), autopct='%1.1f%%', labels=None, startangle=90, colors= color_list, shadow=True, pctdistance=1.12)  

plt.legend(labels=df.index, loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn in 2024') 
plt.axis('equal')
plt.show()
```

```python
#Quiz questions
Query3a="""
SELECT LanguageDesireNextYear, count(*) as count
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count desc limit 10
"""

LanguageDesireNextYear= pd.read_sql_query(Query3a, conn)
print(LanguageDesireNextYear)
```

```python
#Quiz questions
Query3b="""
SELECT respondent,DatabaseWorkedWith FROM DatabaseWorkedWith 
WHERE respondent in (SELECT respondent from (SELECT respondent,count(respondent) as LangCount
FROM DatabaseWorkedWith
GROUP BY Respondent) where LangCount=1)
AND DatabaseWorkedWith='MySQL'
"""

DatabaseDesireNextYear= pd.read_sql_query(Query3b, conn)
print(DatabaseDesireNextYear)
```

```python
#Quiz questions
Query3c="""
select LanguageWorkedWith,count(Respondent) as count
from LanguageWorkedWith  
where LanguageWorkedWith = 'SQL'
"""

DatabaseWorkedWith= pd.read_sql_query(Query3c, conn)
print(DatabaseWorkedWith)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query4 = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master 
where Age between 30 and 35
"""

df = pd.read_sql_query(Query4, conn)
order1 = ['WorkWeekHrs', 'CodeRevHrs']
print(df.groupby('Age')[order1].median())


```

```python
df.groupby('Age')[order1].median().plot.bar(stacked=True)
plt.title('Median hours, by age')
```

```python
Query7 = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master 
where Age between 25 and 30
"""

df = pd.read_sql_query(Query7, conn)
order1 = ['WorkWeekHrs', 'CodeRevHrs']
print(df.groupby('Age')[order1].median())
```

```python
Query9= """
SELECT  Age, ConvertedComp
FROM master
where age between 25 and 30
"""
df9= pd.read_sql_query(Query9, conn)
df9=df9.groupby('Age')['ConvertedComp'].median()
df9.plot()
plt.title('Median ConvertedComp, by age')
print(df)
```
